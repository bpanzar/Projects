```python
def comp_int(P, n, r, y):
    '''     P = int starting principle 
            n = int number of compounding periods per year
            r = float annual interest rate. e.g. 15 for 15% 
            y = int the amount of years
    '''
    return P * (((1 + ((r/100.0)/n)) ** (n*y)))
```

### Time Series Analysis

```python
usa_time = data_df
```

```python
usa_time['Year'] = pd.date_range(start='1/1/1970', end='1/1/2016', periods=47)
```

```python
usa_time.set_index('Year', inplace=True)
usa_time.head()
```

```python
usa_time['Change'] = usa_time.National_Income_Top_1.div(
    usa_time.National_Income_Top_1.shift())
usa_time['Change'].plot(figsize=(20,8))
```

```python
# Autocorrelation of income
#plot_acf(white_noise,lags=15)
plot_acf(usa_time['National_Income_Top_1'],lags=25,title="Top 1%")
plt.show()
```

```python
# Partial Autocorrelation of income
plot_pacf(usa_time['National_Income_Top_1'],lags=25)
plt.show()
```

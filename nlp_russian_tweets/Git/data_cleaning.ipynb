### Data Cleaning

I expect this data to be clean because it has been compiled by researchers at Clemson University and then released by FiveThirtyEight.com, however, I will double check and look for red flags.

I will look for errors first in the categorical features and then in the numerical features.

```python
import pandas as pd
```

```python
#Load data, see how big the dataframe is, look at the first five rows
df = pd.read_csv('/Users/bradpanzarella/Desktop/Fletcher/data/IRAhandle_tweets_1.csv')

print(df.shape)
df.head()
```

### Categorical Features

Let's first see what is in the categorical columns.

```python
df['region'].unique()
```

```python
df['language'].unique()
```

```python
df['account_type'].unique()
```

```python
df['account_category'].unique()
```

Account category seems promising.  Let's see if 'Unknown' is an important category.

```python
df_unk = df[df['account_category'] == 'Unknown']
df_unk.head()
```

```python
df_unk.tail()
```

There does not seem to be much consistency with the 'Unknown' category, but these tweets could be useful.

Now let's see if we can just drop all of the 'NonEnglish' tweets.

```python
df_eng = df[df['account_category'] != 'NonEnglish']
df_eng.head()
```

```python
df_eng['language'].unique()
```

We still have a lot of non-English languages!  Let's look at some.

```python
df_eng[df_eng['language'] != 'English'].head()
```

```python
df_eng[df_eng['language'] != 'English'].tail()
```

Some of these are English but mislablled!  Let's see how many of the other languages we have.

```python
df_eng.groupby('language')['content'].nunique()
```

```python
# Calculate percentage of how many tweets are labelled with 'English'
df_eng[df_eng['language'] == 'English']['content'].count() / df_eng['content'].count()
```

Since the tweets not labelled 'English' represent less than 2% of the total tweets I will let the NLP process take care of the ones that are actually not English.

Let's check the region of the English tweets.

```python
df_eng['region'].unique()
```

```python
df_eng[df_eng['region'] == 'Unknown'].head()
```

Unknown region definitely has value.

```python
df_eng[df_eng['region'] == 'Japan'].head()
```

These strange tweets are labelled as Japan but seem to have some English in them.

```python
df_eng[df_eng['region'] == 'Belarus'].head()
```

These tweets could also be mislabelled.  It looks like 'region' will not be very helpful...

```python
#Fill nan with Unknown
df['region'] = df['region'].fillna('Unknown')
df['region'].unique()
```

We can see that region is unreliable as the following seem to be from the United States.

### Categorical Findings

It looks like the 'account_category' column may be the only really useful column.  Region and language are unreliable.
